```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1
df = pd.read_csv("medical_examination.csv")
```

```python
# 2
df['overweight'] = ((df['weight'] / (df['height']**2)) > 25).astype(int)
df
```

```python
# 3
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
df['gluc'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)
df
```

```python
# 4
df_cat = pd.melt(df, 
    id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 
    'smoke', 'alco', 'active', 'overweight'])
```

```python
df_cat = df_cat.groupby(['cardio', 'variable', 'value'], as_index=False).size().rename(columns={'size': 'total'})
```

```python
df_cat
```

```python
fig = sns.catplot(x="variable", y="total", hue="value", col="cardio", 
        data=df_cat, kind="bar", height=5, aspect=1).figure
```

```python
df_heat = df[ 
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
df_heat
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
```

```python
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", center=0, cmap="coolwarm",
    square=True, linewidth=.5, cbar_kws={"shrink": 5})
```
